```python
def multiply_A_z(A, z):
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            if A[i, j] == 1:
                A[i, j] = z
    return A
```

```python
A3, A4, A5 = multiply_A_z(A3, z), multiply_A_z(A4, z), multiply_A_z(A5, z)
```

```python
def make_polynomial(A):
    E = sympy.eye(A.shape[0])    
    return sympy.expand(sympy.det(E - A))
```

```python
def is_irreducible(polynomial):
    return sympy.factor(polynomial) == polynomial
```

```python
print('Характеристический многочлен для варианта 3:')
print('Неприводимый' if is_irreducible(pol_3) else 'Приводимый')
```

```python
def validate(vector):
    '''
    Проверка, является ли функция функцией Де Брейна
    '''
    coefs = {}
    for i in range(len(vector)):
        coefs[format(i, f'0{int(np.log2(len(vector)))}b')] = [str(vector[i]), 0]
        
    k = '0' * int(np.log2(len(vector)))
    while coefs[k][1] == 0:
        coefs[k][1] = 1
        k = k[1:] + coefs[k][0]
    
    flag = 0
    for key in coefs.keys():
        if coefs[key][1] != 1:
            flag = 1
            
    if flag == 0 and k == '0000':
        return True
    return False
    
    
def find_deBruijn_functions(variables_amount, max_count=None):
    deBruijn_functions = []
    for i in range(2 ** (2 ** variables_amount)):
        vector = format(i, f'0{2 ** variables_amount}b')
        if validate(vector):
            deBruijn_functions.append(vector)
            if max_count and len(deBruijn_functions) == max_count:
                return deBruijn_functions
    return deBruijn_functions
```

```python
def truth_table(vector):
    coefs = {}
    for i in range(len(vector)):
        print(format(i, f'0{int(np.log2(len(vector)))}b'), '|', vector[i])


def functions_graph(vector):
    '''
    Печать графа функции
    '''
    coefs = {}
    for i in range(len(vector)):
        coefs[format(i, f'0{int(np.log2(len(vector)))}b')] = [str(vector[i]), 0]
    
    k = '0' * int(np.log2(len(vector)))
    while coefs[k][1] == 0:
        print(k, '->', end=' ')
        coefs[k][1] = 1
        k = k[1:] + coefs[k][0]
    print(k)
```

```python
def zhegalkin_polynomial(vector):
    anf = [int(i) for i in vector]
    for i in range(len(vector)):
        bin_i = format(i, f'0{int(np.log2(len(vector)))}b')
        print(f'a{bin_i} = {anf[0]}')
        for j in range(len(anf)-1):
            anf[j] = (anf[j] + anf[j + 1]) % 2 
        anf = anf[:-1]
```

```python
def create_Walsh_matrix(shape):
    if shape == (2, 2): 
        return np.array([[1, 1], [1, -1]])
    
    prev_shape = shape[0]//2
    prev_matrix = create_Walsh_matrix((prev_shape, prev_shape))
    current_matrix = np.zeros([2*i for i in prev_matrix.shape])
    for i in range(prev_matrix.shape[0]):
        for j in range(prev_matrix.shape[0]):
            current_matrix[i, j] = prev_matrix[i, j]
            current_matrix[i + prev_shape, j] = prev_matrix[i, j]
            current_matrix[i, j + prev_shape] = prev_matrix[i, j]
            current_matrix[i + prev_shape, j + prev_shape] = -1 * prev_matrix[i, j]
    return current_matrix
        
    
def transform(vector):
    vector = [int(i) for i in vector]
    return np.array([1] * len(vector)) - 2 * np.array(vector)
    

def compute_Walsh_coefs(vector):
    dim = len(vector)
    W = create_Walsh_matrix((dim, dim))
    return W.dot(transform(vector))
```

```python
for i, func in enumerate(functions_list, 1):
    print(f'Функция {i}: {func}')
    print('Таблица истинности функции:')
    truth_table(func)
    print('Цикл вершин графа функции:')
    functions_graph(func)
    print('Многочлен Жегалкина:')
    zhegalkin_polynomial(func)
    walsh_coefs = compute_Walsh_coefs(func)
    print('Коэффициенты Адамара–Уолша:', list(walsh_coefs), '\n\n')
```
